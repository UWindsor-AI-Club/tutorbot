```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1'
}

def get_response_text(input_url, user_headers):
    response = requests.get(input_url, headers=user_headers)
    return response.text


def save_html_output(doc, user_file_name):
    output_path = f'{user_file_name}_doc.html'

    with open(output_path, 'w', encoding='utf-8') as file:
        file.write(doc.prettify()) 

    print(f"Content saved to {output_path}")

def fetch_and_parse_html(input_url, user_file_name, headers=headers):
    output_path = f'{user_file_name}_soup_doc.html'
    
    try:
        response = requests.get(input_url, headers=headers)
        response.raise_for_status()
        doc = BeautifulSoup(response.text, 'html.parser')
        with open(output_path, 'w', encoding='utf-8') as file:
            file.write(doc.prettify())
        return doc
    
    except requests.RequestException as e:
        print(f"An error occurred: {e}")
        return None
```

# GFG 

```python
# GitHub
gfg_url = 'https://www.geeksforgeeks.org'
gfg_topics_url = "python-basics"
gfg_full_url = gfg_url + "/" + gfg_topics_url + "/"

github_topics_contents = get_response_text(gfg_full_url, headers)
soup = BeautifulSoup(github_topics_contents, 'html.parser') 
save_html_output(soup, "gfg_" + gfg_topics_url)
```

```python
with open('gfg_python-basics_doc.html', 'r') as f:
    html_content = f.read()
    soup = BeautifulSoup(html_content, 'html.parser')
```

```python
gfg_h1 = soup.findAll('h1')

for item in gfg_h1:
    print(item.text.strip())


code_blocks = soup.findAll('code', class_='language-python')
```

```python
for code_block in code_blocks:
    pre_block = code_block.find_all('pre')

    for item in pre_block:
        print(item.get_text(strip=True))
        print()
```
